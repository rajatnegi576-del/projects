```python
li=df[df["Distance_km"].notna()]["Distance_km"].unique().tolist()  
#here we are taking all the unique values .
li.remove(0.0)
#now we removed 0.0 also

df.loc[(df["Origin"]==df["Destination"]) & (df["Distance_km"].isna()),"Distance_km"]=0.0
#here we filled the 0.0 values in place of null where destination and origin is same.
#which means the flight has not taken off

#odd is origin destination distance
odd={i: [df[df["Distance_km"]==i] ["Origin"].unique().tolist(), # we are taking unique list of origin
        df[df["Distance_km"]==i]["Destination"].unique().tolist()] #we are taking unique list of destination
        for i in li}
for i,j in odd.items():
    df.loc[(df["Origin"]!=df["Destination"]) &
            (df["Origin"].isin(j[0])) &   #i is iterating in li and j is iterating in origin 
            (df["Destination"].isin(j[1])) &    # and destination
            (df["Distance_km"].isna()),"Distance_km"]=i
#after  the completion of condition we closed the condition with (,) then we made a change in all 
#distance where destination and origin's values are null and filled it.
```

```python
def hms(a):
    h=int(a)
    m=int((a-h)*60)
    s=round((((a-h)*60)-m)*60)
    if s>=60:
        if s==60:
            s=0
        else:
            s=s-60
        m+=1
    if m>=60:
        if m==60:
            m=0
        else:
            m=m-60
        h+=1
    return (f"{h:02d}:{m:02d}:{s:02d}")
df["Duration_hours"]=df["Duration_hours"].apply(lambda x:hms(x)) 
```

```python
def dm(n):
    h=0
    m=int(n)
    s=round((n-int(n))*60)
    while m>=60:
        if m==60:
            m=0
        else:
            m=m-60
        h+=1
    while s>=60:
        if s==60:
            s=0
        else:
            s=s-60
        m+=1
    return (f"{h:02d}:{m:02d}:{s:02d}")
df["Delay_minutes"]=df["Delay_minutes"].apply(lambda x:dm(x))
```
